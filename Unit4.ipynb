## 1. matplotlib 라이브러리 불러오기
---

````python
import matplotlib.pyplot as plt
````

## 2. 기본 그래프 그리기
---

````python
plt.plot([10, 20, 30, 40])
plt.show
````

`plot()` 함수에 입력된 리스트는 y축의 값인것을 확인할 수 있다.<br>
x축 값은 자동으로 0부터 1씩 증가하는 정수로 입력되었다.

````python
plt.plot([1,2,3,4,], [12, 43, 25, 15])
plt.show
````

`plot()` 함수에 입력된 두개의 리스트중 첫번째가 x축의 값이고, 두번째가 y축의 값이다.

plot 함수에 기본 그래프를 그리려면 보통 다음 세단계를 거쳐야 한다.

```python
import matplotlib.pyplot as plt
plt.plot([x축 데이터], [y축 데이터]) #plot() 함수에 데이터 입력하기
plt.show() #그래프 보여주기
```

## 3. 그래프에 옵션 추가하기
---
그래프에 제목, 레이블과 같은 기본옵션을 추가할 수 있다.

### 그래프에 제목넣기
---
제목을 넣는 함수는 `title()` 함수이다.

````python
plt.title('plotting') #제목은 영어로 넣는것이 좋다. 한글을 지원하는 폰트 설정이 이루어지지 않아서 글자가 깨진다.
plt.plot([10, 20, 30, 40])
plt.show()
````

### 그래프에 범례 넣기
---
볌례(legend)를 넣을 수 있다. 보통 두 개 이상의 데이터를 표시할 때 사용한다.

범례를 넣기 위해서는 `plot()` 함수를 사용한다. 함수에 label이라는 속성의 레이블 값으로 원하는 문자열을 넣어주고 그래프를 그리기 전에 `legend()` 함수를 실행시키면 레이블 값이 범례로 나타난다.

````python
plt.title('legend')
plt.plot([10, 20, 30, 40], label='asc') #증가를 의미하는 asc 범례
plt.plot([40, 30, 20, 10], label='desc') #감소를 의미하는 desc 범례
plt.legend()
plt.show()
````

#### 범례의 위치를 지정하기.

```python
plt.legend(loc = 5)
```
`loc` 이라는 속성에 값을 넣어주면 된다.

|2|9|1|
|---|---|---|
|6|10|5, 7|
|3|8|4|

요런 식으로 범례의 위치를 정할 수 있다.

### 그래프 색상 바꾸기
---
원하는 색상으로 바꾸려면 다음과 같이 `color` 속성 추가하면 된다. `color` 속성을 생략하면 자동으로 색상이 설정된다.

````python
plt.title('color') #제목 설정
#그래프 그리기
plt.plot([10, 20, 30, 40], color='skyblue', label='skyblue')
plt.plot([40, 30, 20, 10], 'pink', label='pink')
plt.legend() #범례 표시
plt.show()
````

### 그래프 선 모양 바꾸기
---
`linestyle` 속성에 원하는 선 모양을 지정하면 선모양을 다양하게 바꿀 수 있다. `linestyle` 속성 대신 `ls`라고 작성할 수도 있다.

````python
plt.title('linestyle') #제목 설정
# 빨간색 dashed 그래프
plt.plot([10, 20, 30, 40], color='r', linestyle='--', label='dashed')
# 초록색 dotted 그래프
plt.plot([40, 30, 20, 10], color='g', ls=':', label='dotted')
plt.legend() #범례 표시
plt.show()
````

### 마커 모양 바꾸기
---
`plot()` 함수에 `marker` 속성을 설정하면 선 형태가 아닌 점 형태로 그래프를 그릴 수 있다.

````python
plt.title('marker') #제목 설정
plt.plot([10, 20, 30, 40], 'r.', label='circle') #빨간색 원형 마커 그래프
#초록색 삼각형 마커 그래프
plt.plot([40, 30, 20, 10], 'g^', label='triandgle up')
plt.legend() #범례 표시
plt.show()
````
